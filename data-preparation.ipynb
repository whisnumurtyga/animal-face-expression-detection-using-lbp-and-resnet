### `Import Library`

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, LabelEncoder
```

### `Config`

```python
datasets =  [
                'dogcat-img-16.csv', 
                'dogcat-img-32.csv',
                'dogcat-img-64.csv',
                'dogcat-img-128.csv',
            ]

read_path = './dataset/dog-cat-csv/'

scaler = StandardScaler()
label_encoder = LabelEncoder()
```

### `Main`

```python
tes_df = pd.read_csv(f'{read_path}{datasets[0]}')
tes_df.head()
```

```python
for i in datasets:
    df = pd.read_csv(f"{read_path}{i}")
    df = df.iloc[:,1:]
    
    x = df.iloc[:,1:]
    y = df.iloc[:,[0]]

    y_encoded = pd.DataFrame(label_encoder.fit_transform(y))
    x = pd.DataFrame(scaler.fit_transform(x))
    
    combined_df = pd.concat([x, y_encoded], axis=1)
    combined_df.to_csv(f'./dataset/dogcat-encode/encoded-{i}')
```
