```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, LabelEncoder
```

```python
datasets = ['dataset-resize-16.csv', 
            'dataset-resize-32.csv',
            'dataset-resize-64.csv',
            'dataset-resize-128.csv',
            ]
```

```python
scaler = StandardScaler()

for i in datasets:
    df = pd.read_csv(i)
    df = df.iloc[:,1:]

    y = df.iloc[:,[0]]
    y_encoded = pd.get_dummies(y)
    x = df.iloc[:,1:]

    combined_df = pd.concat([x, y_encoded], axis=1)

    # Melakukan fit transform ke kolom-kolom numerik
    combined_df = pd.DataFrame(scaler.fit_transform(combined_df))
    
    combined_df.to_csv(f'cleaned-{i}')
```
