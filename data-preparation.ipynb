```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, LabelEncoder
```

```python
df = pd.read_csv('dataset-ncdblbpac.csv')
df = df.iloc[:,1:]
df.head()
```

```python
df['0'].value_counts()
```

```python
df.info()
```

```python
# Misalkan 'numeric_cols' adalah daftar fitur-fitur numerik dalam dataset Anda
scaler = StandardScaler()
label_encoder = LabelEncoder()
```

```python
y = df.iloc[:,[0]]
# Misalkan 'y' adalah DataFrame atau Series label kelas Anda
y_encoded = pd.get_dummies(y)
y_encoded
```

```python
X = df.iloc[:,1:]
X


# Normalisasi pada data pelatihan
x_scale = scaler.fit_transform(X)
x_df = pd.DataFrame(x_scale)
x_df
```

```python
# Gabungkan fitur-fitur yang sudah dinormalisasi dengan label
combined_df = pd.concat([x_df, y_encoded], axis=1)
combined_df.to_csv('ncdlbpac-scale.csv')
combined_df.head()
```
