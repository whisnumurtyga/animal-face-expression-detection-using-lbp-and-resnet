```python
import os

# Tentukan direktori awal yang ingin Anda iterasi
root_directory = '../dataset-fix'

# Iterasi semua folder, subfolder, dan file dalam direktori
for root, directories, files in os.walk(root_directory):
    # root adalah path saat ini
    # directories adalah list dari nama-nama subfolder dalam root
    # files adalah list dari nama-nama file dalam root

    # Cetak nama folder saat ini
    folder_name = os.path.basename(root)

    # Iterasi semua subfolder dalam folder saat ini (root)
    for directory in directories:
        subfolder_name = os.path.basename(directory)
        subfolder_path = os.path.join(root, directory)
        # Cetak nama subfolder dengan format yang diminta
        print(f"{folder_name}->{subfolder_name}")

        # Iterasi semua file dalam subfolder saat ini
        for sub_root, sub_directories, sub_files in os.walk(subfolder_path):
            subfolder_files = sub_files
            # Cetak nama file dalam subfolder dengan format yang diminta
            for file in subfolder_files:
                print(f"{folder_name}->{subfolder_name}->{file}")
```
