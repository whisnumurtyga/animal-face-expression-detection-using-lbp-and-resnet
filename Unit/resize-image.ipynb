```python
import cv2
import matplotlib.pyplot as plt
```

```python
def resize_image(image_path, target_size):
    # Baca gambar
    image = cv2.imread(image_path)

    # Resize gambar sesuai dengan ukuran yang ditentukan
    resized_image = cv2.resize(image, (target_size, target_size))

    return resized_image
```

```python
# Nama file gambar Anda
image_path = './img.jpg'

# Ukuran yang diinginkan
sizes = [16, 32, 64, 128]

# Gambar asli
original_image = cv2.imread(image_path)

# Tampilkan gambar asli dan hasil resize
plt.figure(figsize=(10, 15))

# Gambar asli di subplot baris pertama
plt.subplot(3, 2, 1)
plt.title('Original Image')
plt.imshow(cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB))
plt.axis('off')

# Looping untuk resize dengan ukuran yang diinginkan
index = 2
for size in sizes:
    if index <= 6:  # Memastikan hanya melakukan plot hingga subplot ke-6
        resized = resize_image(image_path, size)
        plt.subplot(3, 2, index)
        plt.title(f'Resized to {size}x{size}')
        plt.imshow(cv2.cvtColor(resized, cv2.COLOR_BGR2RGB))
        plt.axis('off')
    index += 1

plt.tight_layout()
plt.show()
```
